```python
import pandas as pd
pd.set_option('display.max_colwidth',200)
pd.set_option('display.max_rows',200)
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import random
from matplotlib.lines import Line2D
```

```python
df=pd.read_csv('E:/4th sem/DAV/data/placements.csv')
print('Shape is: ',df.shape)
print('Columns are: ',df.columns)
```

```python
format_dict={'LPA':'{:.2f}','No_of_students':'{:,}'}
```

```python
df_2019=df[df['Year']==2019]
df_2020=df[df['Year']==2020]
df_2021=df[df['Year']==2021]
df_2022=df[df['Year']==2022]

print('Number of placement offers for 2019 students is: ',df_2019.shape[0])
print('Number of placement offers for 2020 students is: ',df_2020.shape[0])
print('Number of placement offers for 2021 students is: ',df_2021.shape[0])
print('Number of placement offers for 2022 students is: ',df_2022.shape[0])
```

```python
print('Min and max placements packages in 2019 are: ',min(df_2019['LPA']),max(df_2019['LPA']))
mean_2019=sum(df_2019['LPA']*df_2019['No_of_students'])/sum(df_2019['No_of_students'])
print('Mean LPA for 2019 is: ',mean_2019)
print('Min and max placements packages in 2020 are: ',min(df_2020['LPA']),max(df_2020['LPA']))
mean_2020=sum(df_2020['LPA']*df_2020['No_of_students'])/sum(df_2020['No_of_students'])
print('Mean LPA for 2020 is: ',mean_2020)
print('Min and max placements packages in 2021 are: ',min(df_2021['LPA']),max(df_2021['LPA']))
mean_2021=sum(df_2021['LPA']*df_2021['No_of_students'])/sum(df_2021['No_of_students'])
print('Mean LPA for 2021 is: ',mean_2021)
print('Min and max placements packages in 2022 are: ',min(df_2022['LPA']),max(df_2022['LPA']))
mean_2022=sum(df_2022['LPA']*df_2022['No_of_students'])/sum(df_2022['No_of_students'])
print('Mean LPA for 2021 is: ',mean_2022)
```

```python
plt.plot([df_2019.shape[0],df_2020.shape[0],df_2021.shape[0],df_2022.shape[0]],
         label='Number of Offers received',marker='h',color='g')
plt.plot([mean_2019,mean_2020,mean_2021,mean_2022],label='Mean flow in 4 years',marker='o',color='r')
plt.legend(loc=0)
spc=np.arange(len(df['Year'].unique()))
plt.xticks(spc+0,('2019','2020','2021','2022'))
plt.title('Two plots of Mean and Number of offers\n received each year\n')
plt.xlabel('Years')
plt.ylabel('Mean value Or Number of companies')
plt.show()
```

```python
#we can observe that in 2021 more companies have arrived, but the mean is less which says that companies that arrived
#year 2021 did not offer better packages for students in normal case.
#In 2022 the number of companies arrived are less but the mean is higher (better then last 3 years). That indicates the
#companies arrived year 2022 offered better packages to students in normal case.
```

```python
df_off=[]
for i in range(len(df)):
    if 'Off-Campus' in df.loc[i,'Company_name']:
        df_off.append([df.loc[i,'Company_name'],df.loc[i,'LPA'],df.loc[i,'No_of_students']])
off_st=[]
off_lpa=[]
s=[]
for i in range(len(df_off)):
    df_off[i][2]=float(df_off[i][2])
    df_off[i][1]=float(df_off[i][1])
    off_st.append(df_off[i][2])
    off_lpa.append(df_off[i][1])
    s.append(df_off[i][2]*df_off[i][1])

print('Number of students got off campus placements: ',sum(off_st))
print('Highest off campus placements: ',max(off_lpa))
print('Least off campus placements: ',min(off_lpa))
print('Mean LPA of off campus placements:',np.round(sum(s)/sum(off_st),2))

df_on=[]
for i in range(len(df)):
    if 'Off-Campus' not in df.loc[i,'Company_name']:
        df_on.append([df.loc[i,'Company_name'],df.loc[i,'LPA'],df.loc[i,'No_of_students']])
on_st=[]
on_lpa=[]
s=[]
for i in range(len(df_on)):
    df_on[i][2]=float(df_on[i][2])
    df_on[i][1]=float(df_on[i][1])
    on_st.append(df_on[i][2])
    on_lpa.append(df_on[i][1])
    s.append(df_on[i][2]*df_on[i][1])

print('Number of students got On campus placements: ',sum(on_st))
print('Highest On campus placements: ',max(on_lpa))
print('Least On campus placements: ',min(on_lpa))
print('Mean LPA of On campus placements:',np.round(sum(s)/sum(on_st),2))
```

```python
#As mean is effected by outliers let us see the median of each year
lpa_2019=[]
df_2019=df_2019.reset_index(drop=True)
for i in range(len(df_2019)):
  for j in range(df_2019.loc[i,'No_of_students']):
    lpa_2019.append(float(df_2019.loc[i,'LPA']))
print("Median of Placement LPA 2019 is: ",np.median(lpa_2019),
      "and most frequent package in LPA is:",float(stats.mode(lpa_2019)[0]))
lpa_2020=[]
df_2020=df_2020.reset_index(drop=True)
for i in range(len(df_2020)):
  for j in range(df_2020.loc[i,'No_of_students']):
    lpa_2020.append(float(df_2020.loc[i,'LPA']))
print("Median of Placement LPA 2020 is: ",np.median(lpa_2020),
      "and most frequent package in LPA is:",float(stats.mode(lpa_2020)[0]))
lpa_2021=[]
df_2021=df_2021.reset_index(drop=True)
for i in range(len(df_2021)):
  for j in range(df_2021.loc[i,'No_of_students']):
    lpa_2021.append(float(df_2021.loc[i,'LPA']))
print("Median of Placement LPA 2021 is: ",np.median(lpa_2021),
      "and most frequent package in LPA is:",float(stats.mode(lpa_2021)[0]))
lpa_2022=[]
df_2022=df_2022.reset_index(drop=True)
for i in range(len(df_2022)):
  for j in range(df_2022.loc[i,'No_of_students']):
    lpa_2022.append(float(df_2022.loc[i,'LPA']))
print("Median of Placement LPA 2022 is: ",np.median(lpa_2022),
      "and most frequent package in LPA is:",float(stats.mode(lpa_2022)[0]))
```
